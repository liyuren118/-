# 混淆矩阵的应用

```python
# 导入混淆矩阵的函数
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
# 调用函数切割数据集
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3 , random_state = 666)
```

```python
# 建立一个无样本权重模型
clf_noweight=DecisionTreeClassifier(max_depth= 4, random_state=666) # 2个连续特征 分层多少次都可以
```

```python
clf_noweight.fit(X_train, y_train)
```

```python
clf_noweight.score(X_train, y_train)
```

```python
# accracy 准确率
clf_noweight.score(X_test, y_test)
```

```python
y_pred_train = clf_noweight.predict(X_train)
y_pred_test = clf_noweight.predict(X_test)
```

```python
# 训练集的准确率
accuracy_score(y_train , y_pred_train)
```

```python
# 测试集的准确率
accuracy_score(y_test , y_pred_test)
```

```python
# 测试集召回率
recall_score(y_test, y_pred_test)
```

```python
# 绘制混淆矩阵      #感兴趣的是小样本 那个是犯罪犯罪，犯罪清白，清白犯罪，清白清白完全取决于样本大小和业务需要
confusion_matrix(y_test , y_pred_test)
```

```python
# 绘制混淆矩阵
# 改变传递标签的顺序 
# 默认是 0 - 1
confusion_matrix(y_test , y_pred_test , labels=[1,0])
```

```python
# 测试集召回率
recall_score(y_test , y_pred_test )
```

```python
# 精准度
precision_score(y_test, y_pred_test)
```

```python
# 计算召回率
f1_score(y_test , y_pred_test)
```

# 泰坦尼克号海难生存人员预测

## 导入需要的库

```python
# 导入所需要的库
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic
```

```python
titanic.info()
```

```python
titanic.Pclass.value_counts()
```

```python
titanic.Embarked.value_counts()
```

```python
titanic.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
titanic
```

```python
titanic.Sex=titanic.Sex.map({'male':0,'female':1})
```

```python
t= titanic.Embarked.str
```

```python
titanic[['C',"S",'Q']]=t.get_dummies()
```

```python
titanic.drop(columns=['Embarked','S'], inplace= True)
```

```python
titanic
```

## 提取特征数据和标签

```python
X=titanic.iloc[: ,1:]
X
```

```python
y=titanic.iloc[: ,0]
y
```

## 切割训练集和测试集

```python
#先进行切分，再进行填充，不然填充的均值已经涉及到测试集了
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3 , random_state = 666)
```

```python
X_train.shape
X_test.shape
y_train.shape
y_test.shape
```

```python
# 对训练集中的空值进行填充
X_train.Age.fillna(X_train.Age.mean(),inplace=True)
```

## 决策树建模

```python
clf=DecisionTreeClassifier(max_depth=4, random_state= 666)
```

```python
clf.fit(X_train, y_train)
clf.score(X_train ,y_train)
```

```python
# 对测试集进行空值填充
X_test.Age.fillna(X_train.Age.mean() , inplace= True)
```

```python
clf.score(X_test , y_test)
```

## 使用网格搜索

```python
# 确定调整哪些参数, 以及调节范围
# criterion=  基尼  信息熵
# map_depth 1 - 15
# min_samples_leaf 3 - 50
# min_impurity_split 逐渐把这个值变大 2倍 3倍 4倍

# 构造参数字典, 字典中的key,参数名称, 字典中的value 参数变化的序列
d = {"criterion":['gini','entropy'],
    "max_depth":range(1,16),
    "min_samples_leaf":range(3,51),
    "min_impurity_split":np.arange(1e-7 , 1e-6,1e-7 )}  # 起点，终点，步长
```

```python
#实例化模型
clf=DecisionTreeClassifier(random_state=666)
```

```python
grid_search=GridSearchCV(clf,param_grid=d,cv=5,n_jobs=-1,verbose=2)
```

```python
# 使用网格搜索API训练模型
grid_search.fit(X_train,y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

```python
clf.score(X_test,y_test)
```
